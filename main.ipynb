```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

driver = webdriver.Chrome()

def extract_data(page_no):
    url = f"https://www.reed.co.uk/courses/online?page={page_no}" # pagination implemented. 
    driver.get(url) # call the driver. 
    time.sleep(2) 

    titles = driver.find_elements(By.CSS_SELECTOR, "div.top-col > h2 > a") # selects all 25 titles of a single page. 
    providers = driver.find_elements(By.CSS_SELECTOR, "div.top-col > div.provider-name") # selects all 25 providers of a single page. 
    students = driver.find_elements(By.CSS_SELECTOR, "div:nth-child(3) > div.bottom-col > ul > li:nth-child(1) > span") # selects all students amount of a single page. 

    for i in range(len(titles)):
        f = {
            "Title": titles[i].text.strip(), 
            "Provider": providers[i].text.strip() if i < len(providers) else "",
            "Students": students[i].text.strip() if i < len(students) else "",
            "URL": titles[i].get_attribute("href"),
        }
        
        lst.append(f)

    print(f"Page {page_no} scraped successfully.")

lst = []

for i in range(1, 11):
    extract_data(i)

df = pd.DataFrame(lst) # save all datas in pandas dataframe. 
df.to_csv("courses_info.csv", index=False) # index=False to avoid extra index in the column 

driver.quit()
print("\nSaved all data in courses_info.csv") 
```
